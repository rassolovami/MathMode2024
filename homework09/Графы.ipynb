**Задача 4**

```python
import matplotlib.pyplot as plt
```

```python
G = nx.complete_graph(['A', 'B', 'C', 'D', 'E', 'F'])

# Веса
weights = range(1, 16)
for i, (u, v) in enumerate(G.edges):
    G[u][v]['weight'] = weights[i]

# Вершины
pos = nx.spring_layout(G)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=800, node_color="skyblue", font_size=12, font_color="black", edge_color="gray")

# Добавляем веса ребер
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()

# Находим кратчайшие пути между всеми парами вершин
shortest_paths = {}
for u in G.nodes:
    for v in G.nodes:
        if u != v:
            shortest_paths[(u, v)] = nx.shortest_path(G, source=u, target=v, weight='weight')

# Выводим кратчайшие пути между вершинами, из которых исходит больше одного ребра
for (u, v), path in shortest_paths.items():
    if len(path) > 1:
        print(f"Кратчайшие пути между вершинами {u} и {v}:")
        print(path)
```
