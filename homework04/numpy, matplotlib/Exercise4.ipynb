Решить данный интеграл методом трапеций:


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Интегрируемая функция
def integrand(x):
    return np.cos(np.log(x)) / x
```

```python
# Метод трапеций для численного интегрирования
def trapezoidal_rule(func, a, b, n):
    x_values = np.linspace(a, b, n + 1)
    y_values = func(x_values)
    h = (b - a) / n
    result = h * (np.sum(y_values) - 0.5 * (y_values[0] + y_values[-1]))
    return result
```

```python
# Пределы интегрирования
lower_limit = 1
upper_limit = np.exp(1)
```

```python
# Число интервалов для метода трапеций
num_intervals = 1000
```

```python
# Вычисление определённого интеграла
result = trapezoidal_rule(integrand, lower_limit, upper_limit, num_intervals)

print(f"Значение определённого интеграла: {result}")
```

```python
# Построение графика интегрируемой функции
x_values = np.linspace(lower_limit, upper_limit, 1000)
y_values = integrand(x_values)

plt.plot(x_values, y_values, label=r'$\frac{\cos(\ln(x))}{x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('График интегрируемой функции')
plt.show()
```

Решить данный интеграл методом прямоугольников:

```python
# Метод прямоугольников для численного интегрирования
def rectangle_rule(func, a, b, n):
    x_values = np.linspace(a, b, n + 1)
    h = (b - a) / n
    result = h * np.sum(func(x_values))
    return result
```

```python
# Вычисление определённого интеграла
result = rectangle_rule(integrand, lower_limit, upper_limit, num_intervals)
print(f"Значение определённого интеграла: {result}")
```

```python
x_values = np.linspace(lower_limit, upper_limit, 1000)
y_values = integrand(x_values)

plt.plot(x_values, y_values, label=r'$\frac{\cos(\ln(x))}{x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('График интегрируемой функции')
plt.show()
```

Решить данный интеграл методом Симпсона:

```python
# Метод Симпсона для численного интегрирования
def simpson_rule(func, a, b, n):
    x_values = np.linspace(a, b, n + 1)
    h = (b - a) / n
    result = h / 3 * (func(a) + 4 * np.sum(func(x_values[1:-1:2])) + 2 * np.sum(func(x_values[2:-1:2])) + func(b))
    return result
```

```python
# Вычисление определённого интеграла
result = simpson_rule(integrand, lower_limit, upper_limit, num_intervals)
print(f"Значение определённого интеграла: {result}")
```
