1. Создайте график x^2*sin(1/x^2)+x на интервале [−1,1], используя 250 точек. Не забудьте пометить оси!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 250)
```

```python
y = x**2 * np.sin(1 / x**2) + x
```

```python
plt.plot(x, y, label=r'$x^2 \sin\left(\frac{1}{x^2}\right) + x$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы функций 1/(1+x^2) и 1/x^2на интервале [5,25]. (Относительная разница чисел a и b определяется как ∣1−a/b∣)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(5, 25, 100)
```

```python
y1 = 1 / (1 + x**2)
y2 = 1 / x**2
```

```python
relative_difference = np.abs(1 - y1 / y2)
```

```python
plt.semilogy(x, relative_difference, label='Относительная разница')
plt.xlabel('x')
plt.ylabel('Относительная разница')
plt.title('Semilogy график относительной разницы')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список здесь! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn, gamma, beta, erf
```

```python
x = np.linspace(0, 10, 100)
```

```python
y_jn = jn(2, x)  # функция Бесселя второго рода и второго порядка
y_gamma = gamma(x)  # гамма-функция
y_beta = beta(x, 2)  # бета-функция
y_erf = erf(x)  # функция ошибок
```

```python
plt.figure(figsize=(20, 11))

plt.subplot(2, 2, 1)
plt.plot(x, y_jn, label='jn(2, x)')
plt.title('Функция Бесселя второго рода и второго порядка')

plt.subplot(2, 2, 2)
plt.plot(x, y_gamma, label='gamma(x)')
plt.title('Гамма-функция')

plt.subplot(2, 2, 3)
plt.plot(x, y_beta, label='beta(x, 2)')
plt.title('Бета-функция')

plt.subplot(2, 2, 4)
plt.plot(x, y_erf, label='erf(x)')
plt.title('Функция ошибок')

plt.tight_layout()
plt.show()
```

4. Создайте цветовой график sin(x)sin(y) на интервале [−π,π]×[−π,π].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
```

```python
X, Y = np.meshgrid(x, y) # сетка значений
```

```python
f = np.sin(X) * np.sin(Y)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
plot = ax.plot_surface(X, Y, f, cmap='viridis')
fig.colorbar(plot)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$\sin(x)*\sin(y)$')

plt.show()
```
