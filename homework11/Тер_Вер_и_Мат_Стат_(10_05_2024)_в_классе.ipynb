```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform # равномерное распределение
from scipy.stats import norm    # нормальное распределение
```

**Задача 1**

```python
mean = 0
std_dev = 4

# Границы интервала ошибки по модулю
lower_bound = -6
upper_bound = 6

probability = norm.cdf(upper_bound, loc=mean, scale=std_dev) - norm.cdf(lower_bound, loc=mean, scale=std_dev)

print("Вероятность ошибки не превышающей по модулю 6 граммов:", probability)
```

**Задача 2**

```python
from scipy.integrate import quad
```

```python
def pdf_waiting_time(x):
    # Границы интервала для равномерного распределения
    a = 0
    b = 5

    # Плотность вероятности
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0

probability, _ = quad(pdf_waiting_time, 0, 2)
print("Вероятность ожидания не более 2 минут:", probability)
```

**Задача 3**

```python
mean = 161
std_dev = 4

lower_bound = 152
upper_bound = 158

# Вычисляем вероятность попадания в интервал (152;158)
probability_lower = norm.cdf(lower_bound, loc=mean, scale=std_dev)
probability_upper = norm.cdf(upper_bound, loc=mean, scale=std_dev)

fraction_of_suits = probability_upper - probability_lower

print("Доля костюмов для мальчиков с ростом от 152 до 158 см:", fraction_of_suits)
```

```python
import math
```

```python
# Параметры распределения
average_defective_per_week = 100
std_dev_defective_per_week = 12

# Вероятность от 20 до 30 бракованных деталей за неделю
probability_20_to_30 = sum(math.exp(-average_defective_per_week) * (average_defective_per_week ** k) / math.factorial(k) for k in range(20, 31))

# Вероятность хотя бы одной забракованной детали в день
probability_at_least_one_per_day = 1 - math.exp(-average_defective_per_week)

print("Вероятность от 20 до 30 бракованных деталей за неделю:", probability_20_to_30)
print("Вероятность хотя бы одной забракованной детали в день:", probability_at_least_one_per_day)
```

**Задач 5**

```python
import scipy.stats as stats
```

```python
mean_1 = 0
std_dev_1 = 4
```

Для задачи 1

```python
quantile_025_1 = stats.norm.ppf(0.25, loc=mean_1, scale=std_dev_1)
quantile_09_1 = stats.norm.ppf(0.9, loc=mean_1, scale=std_dev_1)

print("Среднее:", mean_1)
print("Дисперсия:", std_dev_1**2)
print("0.25 квантиль", quantile_025_1)
print("0.9 квантиль:", quantile_09_1)
```

```python
x = np.linspace(mean_1 - 4 * std_dev_1, mean_1 + 4 * std_dev_1, 1000)
pdf = stats.norm.pdf(x, mean_1, std_dev_1)
cdf = stats.norm.cdf(x, mean_1, std_dev_1)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, 'bo-', label='PDF')
plt.xlabel('Погрешность в весах (граммы)')
plt.ylabel('Плотность вероятности')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, cdf, 'ro-', label='CDF')
plt.title('CDF')
plt.xlabel('Погрешность в весах')
plt.ylabel('Совокупная вероятность')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
